```python
# Adicionando todos os ID's em uma variável
tweet_ids = df_tweets.tweet_id

# Lista com os ids que não puderam ser recuperados.
tweet_erro_ids = []

# Lista com os tweets retornados pela.
tweet_from_api = []  

# iniciar contador para controlar o tempo entre as chamadas na API do Twitter.
start = time.time()

# iniciar loop entre os ids
for id in tweet_ids:
    try:
        # recupera a informação de um Tweet pelo ID.
        tweet = api.get_status(id)
    
        # Armazena o retorno do tweet.
        tweet_from_api.append({  
            'tweet_id': str(id),
            'favorite_count': tweet.favorite_count,
            'retweet_count': tweet.retweet_count
        })
    except tweepy.TweepError as e:
        print(e)
        # Em caso de erro armazenar para tratamento posterior.
        tweet_erro_ids.append(id)
            
# finalizar contador
end = time.time()
    
# mensagens de conclusão
print('Concluído! O processo levou: ' + str(end - start) + ' segundos')
```

```python
tweet_from_api[0]
```

```python
# Armazenando os dados da API em um arquivo local, tweet_json.txt.
with open('tweet_json.txt', 'w') as file:  
    json.dump(tweet_from_api, file)
```
