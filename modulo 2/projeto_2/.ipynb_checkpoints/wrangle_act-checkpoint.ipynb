```python

# lista com ids de tweets
tweet_ids = df_tweets.tweet_id

# lista para armazenar possíveis tweets com erro
error_ids = []

# criar dictionary para informações dos tweets
tweet_infos = {}

# criar lista para que vai ser usada como linha para replicar e add as infos dos tweets
tweet_infos['tweet'] = []  

# iniciar contador...
print('Iniciando...')
start = time.time()

# iniciar loop entre os ids
for id in tweet_ids:
    try:
        # pegar informação do tweet
        tweet = api.get_status(id)
    
        # jogar no dictionary apenas id, favorite_count e retweet_count
        tweet_infos['tweet'].append({  
            'tweet_id': str(id),
            'favorite_count': tweet.favorite_count,
            'retweet_count': tweet.retweet_count
        })
        print('favorite')
        print(tweet.favorite_count)
    except tweepy.TweepError as e:
        print(e)
        # inserir este id em caso de erro
        error_ids.append(id)
            
# finalizar contador
end = time.time()
    
# mensagens de conclusão
print('Concluído! O processo levou: ' + str(end - start) + ' segundos')
```

```python
# salvar os dados acima em um arquivo local
with open('tweet_json.txt', 'w') as file:  
    json.dump(tweet_infos['tweet'], file)
```
